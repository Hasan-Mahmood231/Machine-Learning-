```python
# split the df in test train
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(df.drop('City',axis=1),df.iloc[:,0],random_state=42,test_size=0.2)
```

- in this data booking and rating are ordinal so we ll use ordinal encoding on them.
- the rating_color are nominal so we ll use one hot encoding on them.

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OrdinalEncoder,OneHotEncoder
transformers = ColumnTransformer(transformers = [
    ('tn1', OrdinalEncoder(categories=[['No','Yes'],['Excellent','Very Good','Good','Average','Poor','Not rated']]),['Has Table booking','Rating text'] ),
    ('tn2', OneHotEncoder(sparse_output=False,drop='first'),['Rating color'] )
],remainder='passthrough')
```

```python
tranformed = transformers.fit_transform(x_train)
```
