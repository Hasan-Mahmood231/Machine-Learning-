```python
import seaborn as sns
import numpy as np
import pandas as pd
from sklearn.preprocessing import FunctionTransformer, PowerTransformer

# Fake skewed data (right skewed income)
data = pd.DataFrame({'Income': np.random.exponential(scale=1000, size=1000)})

# Plot before
sns.kdeplot(data['Income'])

# Log transform
```

```python
log_trf = FunctionTransformer(np.log1p)
data['Income_log'] = log_trf.fit_transform(data[['Income']])

# Power transform
pt = PowerTransformer(method='yeo-johnson')
data['Income_pt'] = pt.fit_transform(data[['Income']])

# Plot after
sns.kdeplot(data['Income_log'])
sns.kdeplot(data['Income_pt'])
```
