```python
# categorical data.
sns.countplot(x = 'mental_health_history',data = df)
plt.show()
```

```python

# Get counts of each anxiety_level
counts = df["anxiety_level"].value_counts()

# Make pie chart
plt.figure(figsize=(6,6))
plt.pie(
    counts.values,              # the counts
    labels=counts.index,        # category labels
    autopct="%1.1f%%",          # show percentages
    startangle=90               # start from top
)

plt.title("Anxiety Level Distribution")
plt.show()
```
